```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.describe()
```

```python
df.drop(["Postal Code"],axis=1,inplace=True)
```

# 

### we are counting the No. of Ship Modes and Ploting its Graph 

```python
ship=pd.value_counts(df['Ship Mode'])
ship_per=pd.value_counts(df['Ship Mode'])/len(df['Ship Mode'])*100
ship_p=pd.DataFrame(ship)
```

```python
ship_p["percentage"]=ship_per
ship_p
```

### as from the data we find 59.7% uses the standard mode of delivery

# 

```python
pd.value_counts(df['Segment'])
```

```python
plt.figure(dpi=125)
plt.title("Distribution of Segment")
plt.pie(pd.value_counts(df['Segment']),labels=("consumers","corporate","home"),
        autopct="%.1f%%",shadow=True,explode=[.07,.08,.09])
plt.show()
```

### Here we can see that majority of the customer is Consumer Segment

# 

```python
country=pd.value_counts(df['State']).sort_values(ascending=False)
summ=pd.value_counts(df['State']).sort_values(ascending=False).sum()
percentage=(country/summ*100)
```

```python
state=pd.DataFrame(country)
state["percentage"]=percentage
state.head()
```

```python
plt.figure(figsize=[15,8])
sns.countplot(x='State',data=df,order=df.State.value_counts().index)
plt.title("Count for order by State")
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

### Best 3 performing stores is from california, New york, Texas with 20%,11%,9% contribution

# 

```python
pd.value_counts(df['Region']).sort_values(ascending=False)
```

```python
plt.figure(dpi=125)
plt.title("Distribution by Region")
plt.pie(pd.value_counts(df['Region']),labels=("West","East","Central","South"),
        autopct="%.1f%%",shadow=True,explode=[.07,.08,.09,.09])
plt.show()
```

            South is among least Performing Region

```python
sub_cat=pd.value_counts(df['Sub-Category']).sort_values(ascending=False)
```

```python
plt.figure(dpi=150)
plt.title('Profit Vs Sales For Segment\n\n we found that more profit in Consumer segment')
sns.scatterplot(x='Sales',y='Profit',hue='Segment',data=df)
plt.grid()
plt.show()
```

```python
df.pivot_table(values='Sales',index='Segment',columns='Discount',aggfunc='median')
```

```python
df.pivot_table(values='Profit',index='Segment',columns='Discount',aggfunc='median')
```

### as the discount is =>30% loss increases 

```python
plt.figure(figsize=[15,10])
plt.title("Profit Vs Discount chart for segment")
sns.histplot(y='Profit',x='Discount',data=df,bins=40,hue="Segment",palette="Dark2",kde=True)
plt.grid()
plt.show()
```

we find that when discount is  Greater than or equal to 30% loss increases as with the discount increases.

```python
plt.figure(figsize=[15,10])
plt.title("Profit Vs Sales chart for category")
sns.histplot(y='Profit',x='Sales',data=df,bins=40,hue="Category",palette="Dark2",kde=True)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=[15,10])
plt.title("Profit Vs Dsicount chart for category")
sns.histplot(y='Profit',x='Discount',data=df,bins=40,hue="Category",palette="Dark2",kde=True)
plt.grid()
plt.show()
```

```python
plt.figure(dpi=125)
plt.title("Profit Vs Dsicount chart")
sns.histplot(y='Profit',x='Discount',data=df,bins=40,hue="Region",palette="Dark2",kde=True)
plt.grid()
plt.show()
```
